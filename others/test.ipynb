```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function u_lambda(y)
def u_lambda(y, lam):
    return (1 / lam) * np.log(1 + np.exp(lam * y))

# Define a range of y values and lambda values
y_values = np.linspace(-50, 50, 500)
lambda_values = [0.1, 0.5]

# Plot the function for different lambda values
plt.figure(figsize=(8, 4))
for lam in lambda_values:
    plt.plot(y_values, u_lambda(y_values, lam), label=f"$\\lambda = {lam}$")

plt.plot(y_values, np.maximum(y_values, 0), label="$\\lambda = +\\infty$", linestyle="--", color="gray")

# Add plot details
plt.xlabel("$y$", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
#plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend(fontsize=10)
plt.grid(alpha=0.5)
plt.savefig(f'smooth_bound.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# plot \lambda e^{-\lambda y}

def lambda_exp(lam, y):
    #return lam * np.exp(-lam * y) / (1 + np.exp(-lam * y))**2
    return lam * 1 / (np.exp(-lam * y) + 2 + np.exp(lam * y))
                      
# Define a range of y values and lambda values
y_values = np.linspace(-100, 100, 500)
lambda_values = [0.1, 1, 5]


# Plot the function for different lambda values
plt.figure(figsize=(8, 4))
for lam in lambda_values:
    plt.plot(y_values, lambda_exp(lam, y_values), label=f"$\\lambda = {lam}$")


plt.plot(y_values, lambda_exp(0, y_values), label="$\\lambda = 0$", linestyle="--", color="gray")

# Add plot details
plt.xlabel("$y$", fontsize=12)
plt.ylabel("$\\lambda e^{-\\lambda y}$", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
#plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend(fontsize=10)
plt.grid(alpha=0.5)
plt.show()
```
